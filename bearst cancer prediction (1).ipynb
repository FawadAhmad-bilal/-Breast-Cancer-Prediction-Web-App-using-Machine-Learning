```python
# 1. Correlation with target
correlation_with_target = df.corr()['diagnosis'].sort_values(ascending=False)
print("Top 15 correlated features:")
print(correlation_with_target[1:16])  # Exclude diagnosis itself

# 2. Feature correlation heatmap (multicollinearity check)
plt.figure(figsize=(20, 16))
correlation_matrix = df.drop('id', axis=1).corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
features_to_drop = [
    'id',  
    'perimeter_mean', 'perimeter_se', 'perimeter_worst',
    'area_mean', 'area_se', 'area_worst',
    # Low importance
    'fractal_dimension_mean', 'fractal_dimension_se', 'fractal_dimension_worst',
    'texture_se', 'smoothness_se', 'symmetry_se'
]
df = df.drop(features_to_drop, axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import  accuracy_score,classification_report
from sklearn.ensemble import GradientBoostingClassifier
```

```python
x=df.drop("diagnosis",axis=1)
y=df["diagnosis"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2,stratify=y)
```

```python
GradientBoostingClassifier().get_params()
```

```python
gbc=GradientBoostingClassifier(
    learning_rate=1,
      max_depth=5,
      n_estimators=100,
    random_state=42)
```

```python
gbc.fit(x_train,y_train)
```

```python
gbc.score(x_test,y_test)
```

```python
gbc.score(x_train,y_train)
```
